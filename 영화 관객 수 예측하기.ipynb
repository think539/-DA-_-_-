```python
dis_rank = all_Movie.groupby('distributor').box_off_num.median().reset_index(name = 'dis_rank_num').sort_values(by = 'dis_rank_num')
dis_rank
```

```python
dis_rank.shape[0]
```

```python
dis_rank['dis_rank_num'] = [i+1 for i in range(dis_rank.shape[0])]
dis_rank
```

```python
all_Movie = pd.merge(all_Movie, dis_rank, how='left')
```
